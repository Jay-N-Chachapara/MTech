```python
class DataSet(Classifier) :
    def __init__(self,dataSet,Xi,Yij,Yi=None,Xj=None) :
        self.X = dataSet.iloc[:,Xi:Xj]
        self.Y = dataSet.iloc[:,Yij]
        encoder = OrdinalEncoder()
        self.X = encoder.fit_transform(self.X)
        self.X_train,self.X_test,self.y_train,self.y_test = train_test_split(self.X,self.Y,test_size=0.2,random_state=4)
        clf = Classifier(self.X_train,self.X_test,self.y_train,self.y_test)
        clf.supportVectorMachine()
        clf.nearestNeighbour()
        clf.decisionTree()
        clf.randomForest()
```
