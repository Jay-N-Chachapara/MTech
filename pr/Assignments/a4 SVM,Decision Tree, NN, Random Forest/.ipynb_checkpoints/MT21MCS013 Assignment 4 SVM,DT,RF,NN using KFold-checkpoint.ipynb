```python
def compute(X,Y) :
    SVM = cross_val_score(SVC(kernel='linear'), X, Y,cv=10)
    DT = cross_val_score(tree.DecisionTreeClassifier(), X, Y,cv=10)
    NN = cross_val_score(KNeighborsClassifier(n_neighbors=3), X, Y,cv=10)
    RF = cross_val_score(RandomForestClassifier(max_depth=64), X, Y,cv=10)
    print('\033[1m'+'Accuracy'+'\033[0m')
    print("SVM",SVM,"\nDecision tree",DT,"\nNearest Neighbour",NN,"\nRandom Forest",RF,sep="\n")
    print('\033[1m'+'\nError'+'\033[0m')
    print("SVM",(1-SVM),"\nDecision tree",(1-DT),"\nNearest Neighbour",(1-NN),
          "\nRandom Forest",(1-RF),sep="\n")
```

### Letter Recognition

```python
letter_recognition_dataSet = pd.read_csv('letter-recognition/letter-recognition.data')
X = letter_recognition_dataSet.iloc[:,1:]
Y = letter_recognition_dataSet.iloc[:,0:1]
letter_recognition_dataSet.describe()
```

```python
compute(X,Y)
```

```python
DummyClass.printConfusionMatrix(X,Y,row=12,col=12)
```
