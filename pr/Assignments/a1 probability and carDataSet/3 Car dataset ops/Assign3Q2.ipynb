```python
"""
Enrollement no : MT21MCS013 Jay 
"""
import numpy as np
import matplotlib.pyplot as plt
#from sklearn.cross_validation import train_test_split
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import seaborn as sns
import scipy
from scipy import stats
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
import category_encoders as ce
df = pd.read_csv(r"https://archive.ics.uci.edu/ml/machine-learning-databases/car/car.data", names = ["buying","maint", "doors", "persons", "lug_boot","safety","class"])
buying=[]
maint = []
door = []
person = []
lug_boot = []
safety = []
class_def = []
for row in df['buying'] :
  if row == 'vhigh' :
    buying.append(4)
  elif row == 'high' :
    buying.append(3)
  elif row == 'med' :
    buying.append(2)
  elif row == 'low':
    buying.append(1)
  else :
    buying.append(0)
for row in df['maint']:
  if row == 'vhigh' :
    maint.append(4)
  elif row == 'high' :
    maint.append(3)
  elif row == 'med' :
    maint.append(2)
  elif row == 'low':
    maint.append(1)
for row in df['doors'] :
  if row == '5more' :
    door.append(5)
  else :
    door.append(int(row))
for row in df['class']:
    if row =='unacc':
        class_def.append('unacceptable')
    elif row == 'acc':
        class_def.append('acceptable')
    elif row == 'good':
        class_def.append('good')
    elif row == 'vgood':
        class_def.append('very good')
    else:
        class_def.append('Failed')
for row in df['persons'] :
  if row == 'more' :
    person.append(5)
  else :
    person.append(int(row))
for row in df['lug_boot'] :
  if row == 'small' :
    lug_boot.append(1)
  elif row == 'med' :
    lug_boot.append(2)
  elif row == 'big' :
    lug_boot.append(3)
  else :
    lug_boot.append(0)
for row in df['safety'] : 
  if row == 'low' :
    safety.append(1)
  elif row == 'med' :
    safety.append(2)
  elif row == 'high' :
    safety.append(3)
  else :
    safety.append(0)

X = df.drop(['class'], axis=1)
y = df['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.33, random_state = 1)

encoder = ce.OrdinalEncoder(cols=['buying', 'maint', 'doors', 'persons', 'lug_boot', 'safety'])
X_train = encoder.fit_transform(X_train)
X_test = encoder.transform(X_test)

print("Average")
print("Buying : ",np.average(buying))
print("Maintenances : ",np.average(maint))
print("Door : ",np.average(door))
print("Person : ",np.average(person))
print("LugBoot : ",np.average(lug_boot))
print("Safety : ",np.average(safety))

print("\nMean")
print("Buying : ",np.mean(buying))
print("Maintenances : ",np.mean(maint))
print("Door : ",np.mean(door))
print("Person : ",np.mean(person))
print("LugBoot : ",np.mean(lug_boot))
print("Safety : ",np.mean(safety))

print("\nVariance")
print("Buying : ",np.var(buying))
print("Maintenances : ",np.var(maint))
print("Door : ",np.var(door))
print("Person : ",np.var(person))
print("LugBoot : ",np.var(lug_boot))
print("Safety : ",np.var(safety))

print("\nMedian")
print("Buying : ",np.median(buying))
print("Maintenances : ",np.median(maint))
print("Door : ",np.median(door))
print("Person : ",np.median(person))
print("LugBoot : ",np.median(lug_boot))
print("Safety : ",np.median(safety))

print("\nStandard Deviation")
print("Buying : ",np.std(buying))
print("Maintenances : ",np.std(maint))
print("Door : ",np.std(door))
print("Person : ",np.std(person))
print("LugBoot : ",np.std(lug_boot))
print("Safety : ",np.std(safety))

print("\nMode")
print("Buying : ",stats.mode(buying))
print("Maintenances : ",stats.mode(maint))
print("Door : ",stats.mode(door))
print("Person : ",stats.mode(person))
print("LugBoot : ",stats.mode(lug_boot))
print("Safety : ",stats.mode(safety),end="\n\n")

print("\nSkewness : ")
print(scipy.stats.skew(X_train, axis=0, bias=True))

print("\nKurtosis :")
scipy.stats.kurtosis(X_train, axis=0, bias=True)
```
