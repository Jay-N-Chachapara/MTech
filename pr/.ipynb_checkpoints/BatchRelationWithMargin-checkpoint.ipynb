```python
#Example 2
y1 = [1,1,1]
y2 = [1,2,2]
y3 = [-1,-5,-7]
y4 = [-1,-8,-10]
w = [1.5,0.5,0.5]
e = 0.5
b = 0.5

def computeR(y,w) :
    r = 0
    for i,j in zip(y,w) :
        r += i*j
    return r
r1 = computeR(y1,w)
r2 = computeR(y2,w)
r3 = computeR(y3,w)
r4 = computeR(y4,w)
print("After wo\n(Values of r1 to r4):\n",r1,r2,r3,r4)
yk = []
wn = []
index = -1
while index < 3 and ((r1 <= b) or (r2 <= b) or (r3 <= b) or (r4 <= b)) :
    index += 1
    yk.clear()
    if r1 <= b :
        yk.append(y1)
    if r2 <= b :
        yk.append(y2)
    if r3 <= b :
        yk.append(y3)
    if r4 <= b :
        yk.append(y4)
    
    if not len(wn) :
        Y = []
        for l,m in zip(w,yk) :
            s2 = 0
            #y2 = 0
            yampli = 0
            for i in m :
                s2 += (b - l*i)
                yampli += i**2
            sigmaMulEp = round((s2/yampli)*e,4)
            for i in m:
                Y.append(sigmaMulEp * i)
        for l,j in zip(w,Y) :
            wn.append(round(l+j,4))
        print("Wn",wn)
    else :
        temp = []
        Y = []
        for l,m in zip(wn,yk) :
            s2 = 0
            #y2 = 0
            yampli = 0
            for i in m :
                s2 += (b - l*i)
                yampli += i**2
            sigmaMulEp = round((s2/yampli)*e,4)
            for i in m:
                Y.append(sigmaMulEp * i)
        for l,j in zip(wn,Y) :
                temp.append(round(l+j,4))
        wn.clear()
        wn = temp
        print("Wn",wn)
    print("W{:}".format(index),wn)
    r1 = computeR(y1,wn)
    r2 = computeR(y2,wn)
    r3 = computeR(y3,wn)
    r4 = computeR(y4,wn)
#     print(r1,r2,r3,r4)
print("Final W :",wn)
print("Wt*y for all\n",r1,r2,r3,r4)
```
