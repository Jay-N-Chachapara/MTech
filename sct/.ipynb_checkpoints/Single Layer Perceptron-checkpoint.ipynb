```python
#Example 2
y1 = [1,1,1]
y2 = [1,2,2]
y3 = [-1,-5,-7]
y4 = [-1,-8,-10]
w = [2,3,4]
e = 0.5

def computeR(y,w) :
    r = 0
    for i,j in zip(y,w) :
        r += i*j
    return r
r1 = computeR(y1,w)
r2 = computeR(y2,w)
r3 = computeR(y3,w)
r4 = computeR(y4,w)
print("After wo\n(Values of r1 to r4):\n",r1,r2,r3,r4)
y = []
wn = []
index = -1
while (r1 < 0) or (r2 < 0) or (r3 < 0) or (r4 < 0):
    index += 1
    y.clear()
    if r1 < 0 :
        y = [(k*e) for k in y1]
    elif r2 < 0 :
        y = [(k*e) for k in y2]
    elif r3 < 0 :
        y = [(k*e) for k in y3]
    else :
        y = [(k*e) for k in y4]
    if not len(wn) :
        for l,m in zip(w,y) :
            wn.append(round(l+m,4))
    else :
        temp = []
        for l,m in zip(wn,y) :
            temp.append(round(l+m,4))
        wn.clear()
        wn = temp
    print("W{:}".format(index),wn)
    r1 = computeR(y1,wn)
    r2 = computeR(y2,wn)
    r3 = computeR(y3,wn)
    r4 = computeR(y4,wn)
#     print(r1,r2,r3,r4)
print("Final W :",wn)
print("Wt*y for all\n",r1,r2,r3,r4)
```

```python
class SingleLayerPerceptron :
    def __init__(self, learning_rate = 0.1, max_iteration = 1000) :
        self.lr = learning_rate
        self.limit = max_iteration
        self.weight = None
        self.bias = None
    
    def activation_function(self,activation) :
        if activation >= 0 :
            return 1
        else :
            return 0
        
    def fit(self,X,y) :
        self.weight = np.zeros_like(X.shape[1])
        self.bias = 0
        for j in range(self.limit) :
            for i in range(X.shape[0]) :
                print(np.dot(self.weight,X[i]) + self.bias)
                y_pred = self.activation_function(np.dot(self.weight,X[i]) + self.bias)
                self.weight = self.weight + self.lr * (y[i] - y_pred) * X[i]
                self.bias = self.bias + self.lr * (y[i] - y_pred)
        print("Training complete")
        print(self.weight)
tryOut = SingleLayerPerceptron()
tryOut.fit(X_train,y_train)
```
